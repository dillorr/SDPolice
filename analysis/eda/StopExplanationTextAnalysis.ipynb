#### Beginning text preprocessing

```python
# library for processing text and cleaning
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
nltk.download("wordnet")

import gensim
from gensim.utils import simple_preprocess
from gensim.models import Phrases
from gensim.models.phrases import Phraser

stop_words = stopwords.words("english")

import en_core_web_sm
nlp = en_core_web_sm.load()
allowed_postags = ["NOUN", "ADJ", "ADV"]

wn = WordNetLemmatizer()

from gensim.models.coherencemodel import CoherenceModel
from gensim.models.ldamodel import LdaModel
from gensim.models.hdpmodel import HdpModel
from gensim.models.wrappers import LdaVowpalWabbit, LdaMallet
from gensim.corpora.dictionary import Dictionary
from numpy import array
```

```python
# Change this to false to run whole dataset
use_sample = False
df2 = df.copy()

explanations = df2.drop(df2[df2['reason_for_stop_explanation'].astype(str).map(len) <= 3].index)
explanations.dropna(subset=['reason_for_stop_explanation'], inplace=True)


if use_sample:
    print("Selecting random training data for model training")
    train_size = 1000
    train_data = df2.sample(train_size)
else:
    train_data = df2

train_data
```

```python
# Combine all the chosen fields into a single text.
# papers = df["reason_for_stop_explanation"] + " " + df["reason_for_stop_detail"]
papers = train_data["reason_for_stop_explanation"]

papers
```

```python
# Processing sentences
def sent_to_words(sentences):
    for sentence in sentences:
        yield (simple_preprocess(sentence, deacc=True))
        
paper_words = sent_to_words(papers)
```

```python
# Clean text and remove stopwords
def remove_stopwords(sentences, lemmatize):
    def stopword_filter(tokens):
        tokens = [token for token in tokens if token not in stop_words]
        if lemmatize:
            return [t.lemma_ for t in nlp(" ".join(tokens)) if t.pos_ in allowed_postags]
        return token
    return [stopword_filter(s) for s in sentences]

paper_words = remove_stopwords(paper_words, lemmatize=True)

paper_words = [e for e in paper_words if e]
paper_words
```

```python
def generate_grams(words, min_count, threshold, verbose=True):
    """
    Generates bigrams and trigrams from input list of words or sentencees.

    Parameters
    ----------
    words : iterable of list of str, optional
        List of words or sentences to generate corpora.
    min_count : float, optional
        Ignore all words and bigrams with total collected count lower than this value.
    threshold : float, optional
        Represent a score threshold for forming the phrases (higher means fewer phrases).
    """
    bigram = Phrases(words, min_count=min_count, threshold=threshold)
    trigram = Phrases(bigram[words], min_count=min_count, threshold=threshold)

    bigram_mod = Phraser(bigram)
    trigram_mod = Phraser(trigram)

    bigrams = [bigram_mod[word] for word in words]
    trigrams = [trigram_mod[bigram_mod[word]] for word in words]

    grams = []

    for bg, tg in zip(bigrams, trigrams):   
        tn = [t for t in tg if t not in bg]
        g = bg + tn
        grams.append(g)
        
    return bigrams, trigrams, grams
                 
                 
# Generate nGrams
bigrams, trigrams, grams = generate_grams(
    words=paper_words, min_count=5, threshold=10
)

bigrams
```

```python
#Build dictionary

dictionary = Dictionary(bigrams)
corpus = [dictionary.doc2bow(text) for text in bigrams]
```

```python
Model = LdaModel(corpus=corpus, id2word=dictionary, num_topics=30)
```

```python
# Build function to visualize the content of the mode
def parse_model_topics(model):

    topics = []

    for index, output in model.show_topics(num_topics=model.num_topics, num_words=15):

        topic_words = [part.split("*")[-1][1:-1] for part in output.split(" + ")]
        topic_coef = [part.split("*")[0] for part in output.split(" + ")]

        topic_words = pd.Series(topic_words)
        topic_coef = pd.Series(topic_coef).astype(float)

        topic_words = topic_words.to_frame("word")
        topic_coef = topic_coef.to_frame("coef")

        topic = topic_coef.join(topic_words).assign(topic_number=index)
        topic = topic.reindex(columns=topic.columns[::-1])

        topics.append(topic)
    

    topics = pd.concat(topics)
    mixture = topics.groupby("topic_number")["word"].apply(list)

    return topics, mixture

output, mixture = parse_model_topics(Model)

display(output)
display(mixture)
```

### Testing with C_V coherence


Coherence measures the relative distance between words within a topic. There are two major types C_V typically 0 < x < 1 and uMass -14 < x < 14. It's rare to see a coherence of 1 or +.9 unless the words being measured are either identical words or bigrams. Like United and States would likely return a coherence score of ~.94 or hero and hero would return a coherence of 1. The overall coherence score of a topic is the average of the distances between words. I try and attain a .7 in my LDAs if I'm using c_v I think that is a strong topic correlation. I would say:
