```python
import dask.dataframe as dd

# # actions taken
# path = r'../policeData/actionsTaken/ripa_actions_taken_datasd.csv'
df1 = dd.read_csv(r'../policeData/actionsTaken/ripa_actions_taken_datasd.csv')

# # basis For Property Seizure
# path = r'../policeData/basisForPropertySeizure/ripa_prop_seize_basis_datasd.csv'
df2 = dd.read_csv(r'../policeData/basisForPropertySeizure/ripa_prop_seize_basis_datasd.csv')

# Merge the csv files.
merged = dd.merge(df1, df2, how='left', on='stop_id')

# Write the output.
merged.compute().to_csv('./dumps/dask/1.csv', index=False)


# # basis for search
# path = r'../policeData/basisForSearchesConducted/ripa_search_basis_datasd.csv'
df3 = dd.read_csv(r'../policeData/basisForSearchesConducted/ripa_search_basis_datasd.csv')

# # disability
# path = r'../policeData/disabilityOfPersons/ripa_disability_datasd.csv'
df4 = dd.read_csv(r'../policeData/disabilityOfPersons/ripa_disability_datasd.csv')

# Merge the csv files.
merged = dd.merge(df3, df4, how='left', on='stop_id')

# Write the output.
merged.compute().to_csv('./dumps/dask/2.csv', index=False)


# # gender
# path = r'../policeData/genderOfPersons/ripa_gender_datasd.csv'
df5 = dd.read_csv(r'../policeData/genderOfPersons/ripa_gender_datasd.csv')

# # property seized
# path = r'../policeData/propertySeized/ripa_prop_seize_type_datasd.csv'
df6 = dd.read_csv(r'../policeData/propertySeized/ripa_prop_seize_type_datasd.csv')

# Merge the csv files.
merged = dd.merge(df5, df6, how='left', on='stop_id')

# Write the output.
merged.compute().to_csv('./dumps/dask/3.csv', index=False)


# # race data
# path = r'../policeData/raceOfPersons/ripa_race_datasd.csv'
df7 = dd.read_csv(r'../policeData/raceOfPersons/ripa_race_datasd.csv')

# # reason for stop data
# path = r'../policeData/reasonForStop/ripa_stop_reason_datasd.csv'
df8 = dd.read_csv(r'../policeData/reasonForStop/ripa_stop_reason_datasd.csv')

# Merge the csv files.
merged = dd.merge(df7, df8, how='left', on='stop_id')

# Write the output.
merged.compute().to_csv('./dumps/dask/4.csv', index=False)


# # stop result
# path = r'../policeData/resultOfStop/ripa_stop_result_datasd.csv'
df9 = dd.read_csv(r'../policeData/resultOfStop/ripa_stop_result_datasd.csv')

# # stop details
# path = r'../policeData/stopDetails/ripa_stops_datasd.csv'
df10 = dd.read_csv(r'../policeData/stopDetails/ripa_stops_datasd.csv')

# Merge the csv files.
merged = dd.merge(df7, df8, how='left', on='stop_id')

# Write the output.
merged.compute().to_csv('./dumps/dask/5.csv', index=False)


```
